## Brazilian E-Commerce Public Dataset

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
df = pd.read_csv('Brazilian E-Commerce Public Dataset by Olist.csv')
df.head()
```

```python
df.shape
```

```python
# Convert date columns to datetime objects
date_cols = ['shipping_limit_date', 'order_purchase_timestamp', 'order_approved_at',
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col])

# Calculate delivery time
df['delivery_time'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

display(df[['order_purchase_timestamp', 'order_delivered_customer_date', 'delivery_time']].head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['delivery_time'].dropna(), bins=50, kde=True)
plt.title('Distribution of Delivery Time (Days)')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Convert date columns to datetime objects
date_cols = ['shipping_limit_date', 'order_purchase_timestamp', 'order_approved_at',
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col])

# Calculate delivery time
df['delivery_time'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

display(df[['order_purchase_timestamp', 'order_delivered_customer_date', 'delivery_time']].head())
```

```python
# Check for missing values
print("Missing values per column:")
display(df.isnull().sum())

# Check for duplicate rows
print("\nNumber of duplicate rows:")
display(df.duplicated().sum())
```

```python
# Display column names and their data types
print("Column names and data types:")
display(df.info())

# Display statistical summary
print("\nStatistical summary:")
display(df.describe())
```

```python
customer_state_counts = df['customer_state'].value_counts().reset_index()
customer_state_counts.columns = ['customer_state', 'unique_customer_count']
print("Unique customers per state:")
display(customer_state_counts)

customer_city_counts = df['customer_city'].value_counts().reset_index()
customer_city_counts.columns = ['customer_city', 'unique_customer_count']
print("\nUnique customers per city:")
display(customer_city_counts)
```

```python
# Analyze purchasing patterns by grouping the data by customer_unique_id

customer_purchase_patterns = df.groupby('customer_unique_id').agg(
    total_orders=('order_id', 'nunique'),
    total_items=('order_item_id', 'sum'),
    total_payment_value=('payment_value', 'sum')
).reset_index()

print("Customer purchasing patterns:")
display(customer_purchase_patterns.head())
```

```python
# Calculate average order value
average_order_value = df['payment_value'].mean()
average_purchase_frequency = customer_purchase_patterns['total_orders'].mean()

print("\nAverage Order Value:", average_order_value)
print("Average Purchase Frequency (per unique customer):", average_purchase_frequency)

print("\nCustomer purchasing patterns (first 5 rows):")
display(customer_purchase_patterns.head())
```

## Analyze product performance

```python
top_selling_categories_items = df.groupby('product_category_name')['order_item_id'].sum().sort_values(ascending=False).head(10)
print("Top 10 Product Categories by Number of Items Sold:")
display(top_selling_categories_items)

top_selling_categories_revenue = df.groupby('product_category_name')['payment_value'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Product Categories by Total Revenue:")
display(top_selling_categories_revenue)

product_category_metrics = df.groupby('product_category_name').agg(
    average_price=('price', 'mean'),
    average_freight_value=('freight_value', 'mean')
).reset_index()

print("\nAverage Price and Freight Value per Product Category:")
display(product_category_metrics.head())
```

```python
# Visualize the relationship between product price and freight value
plt.figure(figsize=(12, 6))
sns.scatterplot(x='price', y='freight_value', data=df)
plt.title('Relationship between Product Price and Freight Value')
plt.xlabel('Price')
plt.ylabel('Freight Value')
plt.show()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['freight_value'], bins=50, kde=True)
plt.title('Distribution of Freight Values')
plt.xlabel('Freight Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Analyze order fulfillment and delivery

```python
#Calculate the average delivery time for each customer state, seller state, and product category
average_delivery_time_customer_state = df.groupby('customer_state')['delivery_time'].mean().reset_index()
average_delivery_time_customer_state = average_delivery_time_customer_state.sort_values(by='delivery_time', ascending=False)
print("Average Delivery Time per Customer State:")
display(average_delivery_time_customer_state.head())

average_delivery_time_seller_state = df.groupby('seller_state')['delivery_time'].mean().reset_index()
average_delivery_time_seller_state = average_delivery_time_seller_state.sort_values(by='delivery_time', ascending=False)
print("\nAverage Delivery Time per Seller State:")
display(average_delivery_time_seller_state.head())

average_delivery_time_product_category = df.groupby('product_category_name')['delivery_time'].mean().reset_index()
average_delivery_time_product_category = average_delivery_time_product_category.sort_values(by='delivery_time', ascending=False)
print("\nAverage Delivery Time per Product Category:")
display(average_delivery_time_product_category.head())
```

```python
#Display the top 10 customer states, seller states, and product categories
top_10_customer_states = average_delivery_time_customer_state.head(10)
print("Top 10 Customer States by Average Delivery Time:")
display(top_10_customer_states)

top_10_seller_states = average_delivery_time_seller_state.head(10)
print("\nTop 10 Seller States by Average Delivery Time:")
display(top_10_seller_states)

top_10_product_categories = average_delivery_time_product_category.head(10)
print("\nTop 10 Product Categories by Average Delivery Time:")
display(top_10_product_categories)

plt.figure(figsize=(12, 6))
sns.barplot(x='customer_state', y='delivery_time', data=top_10_customer_states, palette='viridis')
plt.title('Top 10 Customer States by Average Delivery Time')
plt.xlabel('Customer State')
plt.ylabel('Average Delivery Time (Days)')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='seller_state', y='delivery_time', data=top_10_seller_states, palette='viridis')
plt.title('Top 10 Seller States by Average Delivery Time')
plt.xlabel('Seller State')
plt.ylabel('Average Delivery Time (Days)')
plt.show()

plt.figure(figsize=(14, 7))
sns.barplot(x='product_category_name', y='delivery_time', data=top_10_product_categories, palette='viridis')
plt.title('Top 10 Product Categories by Average Delivery Time')
plt.xlabel('Product Category Name')
plt.ylabel('Average Delivery Time (Days)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Calculate the total payment value for each unique customer
customer_payment_value = df.groupby('customer_unique_id')['payment_value'].sum().reset_index()
print("Total payment value per unique customer:")
display(customer_payment_value.head())

customer_state_counts_df = df.groupby('customer_state')['customer_unique_id'].nunique().reset_index()
customer_state_counts_df.columns = ['customer_state', 'unique_customer_count']
print("\nUnique customers per state:")
display(customer_state_counts_df.head())

customer_city_counts_df = df.groupby('customer_city')['customer_unique_id'].nunique().reset_index()
customer_city_counts_df.columns = ['customer_city', 'unique_customer_count']
print("\nUnique customers per city:")
display(customer_city_counts_df.head())

total_payment_value_state_df = df.groupby('customer_state')['payment_value'].sum().reset_index()
print("\nTotal payment value per customer state:")
display(total_payment_value_state_df.head())

total_payment_value_city_df = df.groupby('customer_city')['payment_value'].sum().reset_index()
print("\nTotal payment value per customer city:")
display(total_payment_value_city_df.head())

category_items_sold = df.groupby('product_category_name')['order_item_id'].sum().reset_index()
print("\nNumber of items sold per product category:")
display(category_items_sold.head())

category_revenue = df.groupby('product_category_name')['payment_value'].sum().reset_index()
print("\nTotal revenue per product category:")
display(category_revenue.head())

product_category_metrics_df = df.groupby('product_category_name').agg(
    average_price=('price', 'mean'),
    average_freight_value=('freight_value', 'mean')
).reset_index()
print("\nAverage price and freight value per product category:")
display(product_category_metrics_df.head())
```

```python
#Calculate the average delivery time per customer state, seller state, and product category
avg_delivery_time_customer_state_df = df.groupby('customer_state')['delivery_time'].mean().reset_index()
print("\nAverage delivery time per customer state:")
display(avg_delivery_time_customer_state_df.head())

avg_delivery_time_seller_state_df = df.groupby('seller_state')['delivery_time'].mean().reset_index()
print("\nAverage delivery time per seller state:")
display(avg_delivery_time_seller_state_df.head())

avg_delivery_time_product_category_df = df.groupby('product_category_name')['delivery_time'].mean().reset_index()
print("\nAverage delivery time per product category:")
display(avg_delivery_time_product_category_df.head())

payment_type_counts_series = df['payment_type'].value_counts()
print("\nPayment type counts:")
display(payment_type_counts_series.head())

avg_installments_per_payment_type_series = df.groupby('payment_type')['payment_installments'].mean()
print("\nAverage installments per payment type:")
display(avg_installments_per_payment_type_series.head())
```

## Implement interactivity

```python
import dash
from dash import dcc
from dash import html
import plotly.express as px
from dash.dependencies import Input, Output
import sys
import threading
import time

# Function to stop the running Dash app (more reliable approach for notebook)
def shutdown_server():
    print("Attempting to shut down the Dash server...")
    # This will raise a SystemExit, which we will catch
    sys.exit(1)

# Check if a Dash app is already running and attempt to shut it down
# This is a basic check and might not be foolproof
if 'app' in globals() and isinstance(app, dash.Dash) and app.server:
    print("Existing Dash app detected. Attempting to shut it down.")
    # Run shutdown in a separate thread to avoid blocking the main thread
    threading.Thread(target=shutdown_server).start()
    # Give the server a moment to shut down
    time.sleep(2)
    print("Shutdown attempt finished.")

# Define external scripts for Plotly.js
external_scripts = [
    'https://cdn.plot.ly/plotly-2.32.0.min.js' # Using a specific Plotly.js version from CDN
]

# Initialize the Dash app
app = dash.Dash(__name__, external_scripts=external_scripts)

# Define the layout
app.layout = html.Div([
    html.H1("Brazilian E-Commerce Public Dataset Dashboard"),

    html.Div([
        html.H2("Customer Demographics"),
        html.Div([
            html.Label("Select Customer State:"),
            dcc.Dropdown(
                id='state-dropdown',
                options=[{'label': state, 'value': state} for state in df['customer_state'].unique()],
                value='SP' # Default value
            )
        ]),
        dcc.Graph(id='customer-state-distribution-filtered'), # Updated ID for filtered plot
        dcc.Graph(id='customer-city-distribution', figure=fig_customer_city_dist)
    ]),

    html.Div([
        html.H2("Purchasing Patterns"),
        html.Div(id='purchase-patterns-summary') # Placeholder for future summary or interactive plot
    ]),

    html.Div([
        html.H2("Product Performance"),
        dcc.Graph(id='top-selling-categories-items', figure=fig_top_selling_items),
        dcc.Graph(id='top-selling-categories-revenue', figure=fig_top_selling_revenue),
        dcc.Graph(id='product-price-freight-scatter', figure=fig_price_freight_scatter),
        dcc.Graph(id='product-price-distribution', figure=fig_price_distribution),
        dcc.Graph(id='freight-value-distribution', figure=fig_freight_distribution)
    ]),

    html.Div([
        html.H2("Order Fulfillment and Delivery"),
        dcc.Graph(id='avg-delivery-time-customer-state', figure=fig_avg_delivery_customer_state),
        dcc.Graph(id='avg-delivery_time_seller_state', figure=fig_avg_delivery_seller_state),
        dcc.Graph(id='avg-delivery_time_product_category', figure=fig_avg_delivery_product_category)
    ]),

    html.Div([
        html.H2("Payment Trends"),
        dcc.Graph(id='payment-type-distribution', figure=fig_payment_type_dist),
        dcc.Graph(id='avg-installments-payment-type', figure=fig_avg_installments_payment_type)
    ]),

    html.Div([
        html.H2("Geographic Trends"),
        dcc.Graph(id='customer-distribution-state-plot', figure=fig_customer_state_dist),
        dcc.Graph(id='customer-distribution-city-plot', figure=fig_customer_city_dist),
        dcc.Graph(id='total-payment-value-state-plot', figure=fig_total_payment_state),
        dcc.Graph(id='total-payment-value-city-plot', figure=fig_total_payment_city)
    ])
])

# Define the callback to update the customer state distribution plot based on dropdown selection
@app.callback(
    Output('customer-state-distribution-filtered', 'figure'),
    Input('state-dropdown', 'value')
)
def update_state_distribution_plot(selected_state):
    filtered_df = df[df['customer_state'] == selected_state]
    filtered_state_counts = filtered_df['customer_city'].value_counts().reset_index()
    filtered_state_counts.columns = ['customer_city', 'unique_customer_count']
    fig = px.bar(filtered_state_counts.head(20), x='customer_city', y='unique_customer_count',
                 title=f'Top 20 Cities in {selected_state} by Number of Unique Customers')
    return fig

# Run the app
if __name__ == '__main__':
    try:
        app.run(debug=True, jupyter_mode='inline')
    except SystemExit:
        print("Dash server potentially shut down.")
    except Exception as e:
        print(f"An error occurred: {e}")
```
